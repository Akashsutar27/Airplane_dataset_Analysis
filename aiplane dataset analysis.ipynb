```python
ax = sns.countplot(x = 'Engine Type' , data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```


#analysis of Price with respect to Engine type with Differnt Plotting using seaborn#



```python
import seaborn as sns
sns.boxplot(x='Engine Type',y='Price', data=df)
```

```python
import seaborn as sns
sns.scatterplot(x='Engine Type',y='Price', data=df)
```

• Jet:

■ Exhibits the widest price range and the highest median price.

■ Shows significant variability in prices.

■ Several outliers suggest some jets are priced well above the general price range.

• Propjet:

■ Has a narrower price range than jets.

■ Lower median price compared to jets.

■ Outliers present above and below the main price range, indicating notable price deviations.

• Piston:

■ Features the lowest median price and the smallest interquartile range (IQR), suggesting less price variability.

■ An outlier indicates the presence of at least one piston-engine aircraft with a price much higher than typical for its category.

We can see here the most price is on jet (that's supposed to be true)
